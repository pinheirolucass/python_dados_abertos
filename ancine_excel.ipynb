## Análise de dados de Filmes Nacionais e Estrangeiros exibidos no Brasil de 2009 à 2018
fonte: Ancine , disponível em: http://oca.ancine.gov.br

```python
#importando biblioteca pandas e chamando de pd

#plotar gráficos inline

import pandas as pd

%matplotlib inline

```

```python
#criar DataFrame e pular 2 primeiras linhas do arquivo e pulando linhas do fim do arquivo

df = pd.read_excel('ancine_2120.xlsx',skiprows=2,skipfooter=16)

df.head()
```

```python
df.tail()
```

```python
#verificar os tipos das colunas

df.dtypes
```

```python
#identificar problema na coluna de data de lançamento

df['Data de lançamento'].value_counts()
```

```python
#exibir os dados dos filmes que estão com a data "relançamento"

df[df['Data de lançamento'] =='Relançamento']
```

## Tratar os dados incorretos / faltantes / divergentes

```python
# criar um DataFrame novo e apagar as 175 linhas que tem a data inválida

df_novo = df[df['Data de lançamento'] != 'Relançamento']

df.head()
```

```python
#converter o campo data de lançamento para o tipo data

df_novo['Data de lançamentno'] = pd.to_datetime(df_novo['Data de lançamento'])
```

```python
df_novo.dtypes
```

```python
#Quais os anos tiveram mais filmes lançados

df_novo['Data de lançamento'].dt.year.value_counts()
```

```python
#plotando gráfico

df_novo['Data de lançamento'].dt.year.value_counts().plot.bar()
```

```python
#filme com mair bilheteria

df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].max()]
```

```python
#top 3 bilheterias

df_novo.nlargest(3,'Renda (R$) no ano de exibição')
```

```python
#filme com menor bilheteria


df_novo[df_novo['Renda (R$) no ano de exibição'] == df_novo['Renda (R$) no ano de exibição'].min()]
```

```python
#quantidade de filmes brasileiros e estrangeiros

df_novo['Nacionalidade da obra'].value_counts()
```

```python
df_novo[df_novo['Nacionalidade da obra'] == 'Brasileira']
```

```python
#maior renda de filmes por gênero
df_novo.loc[df_novo.groupby('Gênero')['Renda (R$) no ano de exibição'].idxmax()]
```
