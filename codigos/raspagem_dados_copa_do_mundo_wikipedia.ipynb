## Análise dos dados da copa do mundo

Com base nos dados disponibilizados na Wikipedia, vamos obter os dados do público pagante em cada final.


Link: https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA

```python
## importando bibliotecas

import pandas as pd
import requests

#url
url = 'https://pt.wikipedia.org/wiki/Copa_do_Mundo_FIFA'

#requisitar a página

resposta = requests.get(url)

#exibir a resposta em formato HTML

table = pd.read_html(resposta.text)

#olhar o resultado

table
```

```python
#pegar apenas a tabela com o maior público pagante por final
#12° tabela

df = table[11]

df.head()

#caso os nomes das colunas estejam na primeira linha da tabela usar o comando abaixo

#df.columns = df.iloc[0]

#apagar a primeira linha 
#df.drop(df.index[0], inplace=True)
```

```python
df.info()
```

```python
#caso precise converter o tipo de dado da coluna público pagante

#tirar o ponto dos números 
df['Público pagante'] = df['Público pagante'].astype(int)

##df['Público pagante'] = df['Público pagante'].str.replace('.','').astype(int)

```

```python
#gráfico com a evolução do público pagante

#importar uma biblioteca nova para gráficos

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
#exibindo gráfico de linhas com marcação em cada ponto

sns.pointplot(x='Ano',y='Público pagante', data=df)

#ajustando legendas

plt.xticks(rotation=65)

plt.show()
```
