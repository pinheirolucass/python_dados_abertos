# Análise de Dados de eleitores brasileiros 2018

```python
#importando bibliotecas

import pandas as pd
```

```python
#carregar o arquivo csv

df = pd.read_csv('eleitorado_municipio_2018.csv',sep=';',encoding='latin1')

df.head()
```

```python
#tipo e quantidade dos dados

df.info()
```

```python
#quantidade de eleitores por UF

df.groupby('SG_UF')['QTD_ELEITORES'].sum()
```

```python
#Cidades com maior número de eleitores deficientes

df.nlargest(5,'QTD_ELEITORES_DEFICIENTE')
```

```python
print('Eleitoras',df['QTD_ELEITORES_FEMININO'].sum())

print('Eleitores',df['QTD_ELEITORES_MASCULINO'].sum())

print('Não Informado',df['QTD_ELEITORES_NAOINFORMADO'].sum())
```

```python
#Criar variáveis para calcular percentual

tot_eleitores = df['QTD_ELEITORES'].sum()
tot_feminino = df['QTD_ELEITORES_FEMININO'].sum()
tot_masculino = df['QTD_ELEITORES_MASCULINO'].sum()
tot_nao = df['QTD_ELEITORES_NAOINFORMADO'].sum()
```

```python
#mostrando o resultado de eleitores em porcentagem

print('Eleitoras',((tot_feminino/tot_eleitores)*100).round(2))

print('Eleitores',((tot_masculino/tot_eleitores)*100).round(2))

print('Não Informado', ((tot_nao/tot_eleitores)*100).round(2))
```

```python
#quantas cidades possuem mais mulheres eleitoras do que homens

df[df['QTD_ELEITORES_FEMININO'] > df['QTD_ELEITORES_MASCULINO']].count()
```

```python
#Criar uma coluna nova para indicar a relação fem/masc

df['RELACAO_FEM_MASC'] = df['QTD_ELEITORES_FEMININO'] / df['QTD_ELEITORES_MASCULINO']
```

```python
#Municipios com a amaior relação fem/masc

df.nlargest(5,'RELACAO_FEM_MASC')
```

```python
#municipios com a menor relação fem/masc

df.nsmallest(5,'RELACAO_FEM_MASC')
```

```python
#quantidade de eleitores sem biometria por UF

df.groupby('SG_UF')['QTD_ELEITORES_SEMBIOMETRIA'].sum()
```

```python
#criar um dataframe somente com municípios brasileiros

df_brasil = df[df['CD_PAIS'] == 1].copy()

df_brasil.head()
```

```python
#quantas cidades possuem mais mulheres eleitoras do que homens

df_brasil[df_brasil['QTD_ELEITORES_FEMININO'] > df_brasil['QTD_ELEITORES_MASCULINO']].count()
```

```python
#Municipios brasileiros com a amaior relação fem/masc

df_brasil.nlargest(5,'RELACAO_FEM_MASC')
```

```python
#Municipios brasileiros com a menor relação fem/masc

df_brasil.nsmallest(5,'RELACAO_FEM_MASC')
```

```python
#municipios com maior número de pessoas com 16 anos

df_brasil.nlargest(5,'QTD_ELEITORES_16')
```

```python
#Municipios com a maior quantidade de pessoas que optaram por nome social

df_brasil.nlargest(5,'QUANTITATIVO_NOMESOCIAL')
```
